```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
import matplotlib.pyplot as plt

df=pd.read_csv('ababab.csv',index_col='date' ,parse_dates=True)
df=df.dropna()
df.head()
```

```python
print('Shape of data',df.shape)
```

```python
features = ['meantempm' , 'maxtempm','meanpressurem_1']
data = df[features].values
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 8))

# Plotting meanpressurem_1
axes[0].plot(df.index, df['meanpressurem_1'])
axes[0].set_title('Mean Pressure 1')
axes[0].set_ylabel('Pressure')

# Plotting meantempm
axes[1].plot(df.index, df['meantempm'])
axes[1].set_title('Mean Temperature')
axes[1].set_ylabel('Temperature')

# Plotting precipm_1
axes[2].plot(df.index, df['precipm_1'])
axes[2].set_title('Precipitation')
axes[2].set_ylabel('Precipitation')

# Adjust layout
plt.tight_layout()
plt.xlabel('Date')
plt.show()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)
```

```python
sequence_length = 30
```

```python
# Prepare data for training
X, y = [], []
for i in range(len(data_scaled) - sequence_length):
    X.append(data_scaled[i:i+sequence_length, :])
    y.append(data_scaled[i+sequence_length, :])  

X, y = np.array(X), np.array(y) 
```

```python
# Split the dataset into training and testing sets
split = int(0.8 * len(data_scaled))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = Sequential()
model.add(SimpleRNN(units=30, input_shape=(X.shape[1], X.shape[2]), activation="relu"))
model.add(Dense(7, activation="relu"))
model.add(Dense(3))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
history = model.fit(X, y, epochs=50, batch_size=30, validation_split=0.1, verbose=1)
```

```python
# Make predictions on the test set
test_predictions_scaled = model.predict(X_test)

# Inverse transform the entire test set for plotting
X_test_inverse = scaler.inverse_transform(X_test[:, -1, :])
y_test_inverse = scaler.inverse_transform(y_test)
test_predictions = scaler.inverse_transform(test_predictions_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# Plot the results
for i in range(2):
    plt.plot(df.index[split+sequence_length:], y_test_inverse[:, i], label='Actual ' + features[i])
    plt.plot(df.index[split+sequence_length:], test_predictions[:, i], label='Predicted ' + features[i], linestyle='dashed')

plt.ylabel('Values')
plt.xlabel('Date')
plt.legend()
plt.title('RNN Multivariate')
plt.tight_layout()
plt.show()
```

```python
# Plot actual vs predicted for each feature as subplots
plt.figure(figsize=(12, 8))

# Plot the results
for i in range(3):
    plt.subplot(3, 1, i+1)
    plt.plot(df.index[split+sequence_length:], y_test_inverse[:, i], label='Actual ' + features[i])
    plt.plot(df.index[split+sequence_length:], test_predictions[:, i], label='Predicted ' + features[i], linestyle='dashed')
    plt.ylabel(features[i])
    plt.legend()

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
rmse_per_feature = np.sqrt(mean_squared_error(y_test_inverse, test_predictions, multioutput='raw_values'))

# Print RMSE for each feature
for i, feature in enumerate(features):
    print("RMSE for", feature, ":", rmse_per_feature[i])
```

```python
# Plot epochs vs loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
